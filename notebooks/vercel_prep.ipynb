```python
import json
import ast
import math

def row_to_payload(row):
    """
    Convert a DataFrame row to the API payload format
    """
    # Parse recommendations string to list of lists
    try:
        # Use ast.literal_eval to safely parse the string
        recs = ast.literal_eval(row['recommendations'])
        # Convert tuples to lists: [('name', nova, code)] -> [['name', nova, code]]
        recommendations = [[rec[0], rec[1], rec[2]] for rec in recs]
    except:
        recommendations = []
    
    # Handle NaN values
    def clean_value(val):
        if isinstance(val, float) and math.isnan(val):
            return None
        return val
    
    payload = {
        "cluster_id": int(row['cluster_id']),
        "explain_nova": str(row['explain_nova']),
        "nova_score": int(row['nova_group']),
        "processing_time": "0.001s",  # Precomputed = instant
        "product_macros": {
            "additive_density": float(row['additive_density']),
            "additives_n": float(row['additives_n']),
            "allergens_en": clean_value(row['allergens_en']),
            "carbohydrates_100g": float(row['carbohydrates_100g']),
            "categories_tags": str(row['categories_tags']),
            "code": int(row['code']),
            "fiber_100g": float(row['fiber_100g']),
            "ingredient_count": int(row['ingredient_count']),
            "ingredients_text": str(row['ingredients_text']),
            "monounsaturated-fat_100g": float(row['monounsaturated-fat_100g']),
            "nova_group": int(row['nova_group']),
            "nutriscore_score": float(row['nutriscore_score']),
            "polyunsaturated-fat_100g": float(row['polyunsaturated-fat_100g']),
            "product_name": str(row['product_name']),
            "saturated-fat_100g": float(row['saturated-fat_100g']),
            "sodium_per_calorie": float(row['sodium_per_calorie']),
            "sugar_fiber_ratio": float(row['sugar_fiber_ratio']),
            "sugars_100g": float(row['sugars_100g'])
        },
        "product_name": str(row['product_name']),
        "recommendations": recommendations,
        "success": True
    }
    
    return payload


# Test with one row
test_row = df.iloc[0]
payload = row_to_payload(test_row)

# Pretty print to verify
print(json.dumps(payload, indent=2))
```

```python
all_payloads = {}

for idx, row in df.iterrows():
    product_name = row['product_name']
    payload = row_to_payload(row)
    all_payloads[product_name] = payload

# Save to JSON file
with open('frontend/public/demo_snapshots/precomputed_payloads.json', 'w') as f:
    json.dump(all_payloads, f, indent=2)

print(f"Created precomputed_payloads.json with {len(all_payloads)} products")
```
