```python
code_list = demo_df['code'].values
url_list = dict()
for cd in code_list:
    code = str(cd) 
    response = requests.get(f"https://world.openfoodfacts.org/api/v0/product/{code}.json")
    data = response.json()
    prod = data['product']
    url = prod.get('image_url')
    if url is not None:
        url_list[code] = data['product']['image_url']
    
```

```python
save_dir = '../frontend/demo_snapshots/'


for k, v in url_list.items():
    try:
        response = requests.get(v, stream=True)  # stream=True for large images
        if response.status_code == 200:
            ext = v.split('.')[-1]  # get extension from URL
            file_path = save_dir + k + '.' + ext
            with open(file_path, 'wb') as f:
                for chunk in response.iter_content(1024):
                    f.write(chunk)
            print(f"Saved {file_path}")
        else:
            print(f"Failed to download {v}")
    except Exception as e:
        print(f"Error downloading {v}: {e}")
```

```python
# Install if not already
# pip install transformers torch

from transformers import pipeline

# Use a LLaMA-based model hosted on Hugging Face
# You can pick any text-generation model that doesnâ€™t require authentication
model_name = "chavinlo/alpaca-native"  # smaller and free to use for short text generation

generator = pipeline("text-generation", model=model_name)

# Suppose meta_demo_df is your DataFrame
# We'll loop over rows to make prompts and store outputs
nova_explanations = []

for idx, row in meta_demo_df.iterrows():
    product_name = row['product_name']
    nova_score = row['nova_group']
    ingredients = row['ingredients_text'] if 'ingredients_text' in row else ""
    additives = row['additives_n'] if 'additives_n' in row else 0
    
    # Build a short prompt
    prompt = (f"Explain in one simple line why the product '{product_name}' "
              f"has NOVA score {nova_score}, given ingredients: {ingredients}, "
              f"and number of additives: {additives}.")
    
    # Generate text
    output = generator(prompt, max_length=50, do_sample=True, temperature=0.7)
    
    # Extract generated text and store
    generated_text = output[0]['generated_text']
    nova_explanations.append(generated_text)

# nova_explanations now contains 19 AI-generated one-liners
for i, line in enumerate(nova_explanations, 1):
    print(f"{i}: {line}")
```
