```python
code_list = demo_df['code'].values
url_list = dict()
for cd in code_list:
    code = str(cd) 
    response = requests.get(f"https://world.openfoodfacts.org/api/v0/product/{code}.json")
    data = response.json()
    prod = data['product']
    url = prod.get('image_url')
    if url is not None:
        url_list[code] = data['product']['image_url']
    
```

```python
ans = [int(i) for i in code_list if str(i) not in url_list.keys()]
demo_df[demo_df['code'].isin(ans)][['code', 'product_name']]
```

```python
save_dir = '../frontend/demo_snapshots/'


for k, v in url_list.items():
    try:
        response = requests.get(v, stream=True)  # stream=True for large images
        if response.status_code == 200:
            ext = v.split('.')[-1]  # get extension from URL
            file_path = save_dir + k + '.' + ext
            with open(file_path, 'wb') as f:
                for chunk in response.iter_content(1024):
                    f.write(chunk)
            print(f"Saved {file_path}")
        else:
            print(f"Failed to download {v}")
    except Exception as e:
        print(f"Error downloading {v}: {e}")
```
